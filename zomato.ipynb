```python
plt.figure(figsize=(18,15))
# Setting up size of heatmap
sns.heatmap(data=corr, annot=corr*100,fmt='.0f')
# Giving the correlation as a data to heat map
# corr*100 means taking two integer values after decimal point
# .0f means after decimal point we want nothing that is no number and it also round off the number
plt.show()
```
